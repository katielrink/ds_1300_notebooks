```python
#Imports
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

%config InlineBackend.figure_format ='retina'
```

```python
#Creating Dataset
df = pd.read_csv('../data/state_facts.tsv',delimiter="\t")

#Creating Plot - Showing side by side land and water for each State
ax = df.plot.bar(x='State', y=['Area_land','Area_water'], stacked=True, title='Land vs. Water for Each State', color = ['Green', 'Blue'], figsize=(20,10))

#Add legend
plt.legend(['Land Area', 'Water Area'], loc='upper right', title='Area Type')

#Add axes
ax.set_xlabel("State")
ax.set_ylabel("Acrs. of Land")

#Add description of plot, then description of decisions
txt="This plot is a stacked bar graph that shows both the land area, and the water area at each state in the USA\n"
txt+="I chose to make the plot a bar graph because the data is discrete to each state, and by placing each value in a stacked form you get the total area as well as it's individual breakup\n"
txt+="The colors are corresponding to the colors generally associated with land and water. And the axis includes the units and what data is shown"
plt.figtext(0.5, -.2, txt, wrap=True, horizontalalignment='center', fontsize=12)
```

```python
#Remove large data points to I can see the small ones
subset = df.loc[(df['Area_land'] < 500000) & (df['Area_water'] < 500000)]

#Creating Plot - Showing side by side land and water for each State
ax = subset.plot.bar(x='State', y=['Area_land','Area_water'], stacked=True, title='Land vs. Water for Each State', color = ['Green', 'Blue'], figsize=(20,10))

#Add legend
plt.legend(['Land Area', 'Water Area'], loc='upper right', title='Area Type')

#Add axes
ax.set_xlabel("State")
ax.set_ylabel("Acrs. of Land")
```
